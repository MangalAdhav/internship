```python
#Q.2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
#results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
#scrape all the products available under that product name. Details to be scraped are: "Brand 
#Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
#“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 
```

```python
# fetching URLs to open the pages
urls = []          # empty list
for i in range(0,3):      # for loop to scrape 3 pages
    page_url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in page_url:
        urls.append(i.get_attribute("href"))
        next_btn = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        time.sleep(3)
```

```python
urls
```

```python
len(urls)
```

```python
# making empty list and fetching required data
brand_name = []
product_name = []
prices = []
exchange = []
expect_delivery = []
availability = []
product_url=[]

for i in urls:
    
    driver.get(i)
    
    
    #fetching brand name 
    try:
        brand = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[20]/div[1]/div/div[1]/div[1]/div[3]/div[2]/div/span/a')
        brand_name.append(brand.text)
    except NoSuchElementException:
        brand_name.append('-')
    
    
    # fetching Name of the Product
    try:
        product = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[4]/div[1]/div/h1/span')
        product_name.append(product.text)
    except NoSuchElementException:
        product_name.append('-')
            
            
           

    #fetching price of the product    
    try:
        price =driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[1]/div/div')
        prices.append(price.text)
    except NoSuchElementException:
        prices.append('-')
        
        
    #fetching return/exchange
    try:
        exch = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[20]/div[1]/div/div[1]/div[1]/div[4]/div[2]/span/a/span')
        exchange.append(exch.text)
    except NoSuchElementException:
        exchange.append('-')
        

    #fetching expected delivery
    try:
        delivery = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[3]/div[10]/div[1]/div/div/div')
        expect_delivery.append(delivery.text)
    except NoSuchElementException:
        expect_delivery.append('-')
        

    #fetching availability information
    try:
        available = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[6]/div/div[1]/span')
        availability.append(available.text)
    except NoSuchElementException:
        availability.append('-')
        
    #url details
    try:
        url = driver.find_element(By.XPATH,'//a[@class="ssf-share-trigger"]')
        product_url.append(url.text)
    except NoSuchElementException:
        product_url.append('-')
```

```python
print(len(brand_name),
len(product_name),
len(prices),
len(exchange),
len(expect_delivery),
len(availability),
len(product_url))
```

```python
# Creating the DataFrame for the scraped data

df = pd.DataFrame({'Brand_name':brand_name,'Name_of_the_product':product_name,'Price_of_the_product':prices,'Return/Exchange':exchange,'Expected_delivery':expect_delivery,'Availability':availability,'Product_url':product_url})
df
```

```python
#saving the data in csv
df.to_csv("df.csv")
```

```python
#Q4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
#and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand 
#Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
#“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
#details is missing then replace it by “- “. Save your results in a dataframe and CSV. 
```

```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import requests
from selenium.webdriver.common.by import By
from selenium import webdriver
```

```python
# connecting to the webdriver


driver=webdriver.Chrome()
driver.get('https://www.flipkart.com/')
```

```python
#user will enter key to search
user_input=input('Enter the product(smartphone_name) that you want to search : ')
```

```python
# searching for user input
search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
search.send_keys(user_input)
```

```python
s_btn = driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
s_btn.click()
```

```python
# fetching 1st page of URLs of smartphone
main_page_url = []
urls = driver.find_elements(By.XPATH,"//a[@class='_1fQZEK']")
for url in urls:
    main_page_url.append(url.get_attribute('href'))
```

```python
len(main_page_url)
```

```python
#creating empty list

brand_name=[]
smartphone_name=[]
colour=[]
RAM=[]
storage_ROM=[]
primary_camera=[]
secondary_camera=[]
display_size=[]
battery_capacity=[]
price=[]
product_url=[]


# scraping data from each url of page 1
for i in main_page_url:
    
    driver.get(i)
    
    
    
    # fetching Name of the smartphone
    try:
        phone = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
        smartphone_name.append(phone.text)
    except NoSuchElementException:
        smartphone_name.append('-')
            
            
           

    #fetching colour of the colour   
    try:
        color =driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        colour.append(color.text)
    except NoSuchElementException:
        colour.append('-')
        
        
    #fetching RAM details
    try:
        ram = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[5]/div/div[2]/div[1]/div[4]/table/tbody/tr[2]/td[2]/ul/li')
        RAM.append(ram.text)
    except NoSuchElementException:
        RAM.append('-')
        

    #fetching storage ROM details
    try:
        rom = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div[1]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        storage_ROM.append(rom.text)
    except NoSuchElementException:
         storage_ROM.append('-')
        

    #fetching primary camera
    try:
        camera1 = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[2]/td[2]/ul/li')
        primary_camera.append(camera1.text)
    except NoSuchElementException:
        primary_camera.append('-')
        
    #fetching secondary camera
    try:
        camera2 = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        secondary_camera.append(camera2.text)
    except NoSuchElementException:
        secondary_camera.append('-')
        
        
    #fetching display size details
    try:
        display = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li')
        display_size.append(display.text)
    except NoSuchElementException:
        display_size.append('-')
        
        #fetching battery capacity details
    try:
        battery = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div[1]/div[9]/table/tbody/tr/td[2]/ul/li')
        battery_capacity.append(battery.text)
    except NoSuchElementException:
        battery_capacity.append('-')
        
        
        #fetching prices
    try:
        prices = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[4]/div[1]/div/div[1] ')
        price.append(prices.text)
    except NoSuchElementException:
        price.append('-')
    
        
    
    #fetching product_url
    try:
        url = driver.find_element(By.XPATH,'//a[@class="ssf-share-trigger"]')
        product_url.append(url.text)
    except NoSuchElementException:
        product_url.append('-')
    
```

```python
#5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps. 
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import requests
from selenium.webdriver.common.by import By
from selenium import webdriver
import re
```

```python
# connecting to the webdriver
driver=webdriver.Chrome()
driver.get('https://www.google.co.in/maps')
```

```python
# entering the city name in search bar
user_input = input('Enter City name to search on google map : ')
search = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/form/input')
search.click()
```

```python
#sending keys to find cities
search.send_keys(user_input)
```

```python
#clicking on search button 
search_btn = driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/div[8]/div[3]/div[1]/div[1]/div/div[2]/div[1]/button")
search_btn.click()
```

```python
try:
    url_str = driver.current_url
    print("URL Extracted: ", url_str)
    latitude_longitude = re.findall(r'@(.*)data',url_str)
    if len(latitude_longitude):
        lat_lng_list = latitude_longitude[0].split(",")
        if len(lat_lng_list)>=2:
            latitude = lat_lng_list[0]
            longitude = lat_lng_list[1]
        print("Latitude = {}, Longitude = {}".format(latitude, longitude))
except Exception as e:
        print("Error: ", str(e))
        
```
