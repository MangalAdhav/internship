```python
# fetching URLs to open the pages
urls = []          # empty list
for i in range(0,3):      # for loop to scrape 3 pages
    page_url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in page_url:
        urls.append(i.get_attribute("href"))
        next_btn = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        
```

```python
# making empty list and fetching required data
brand_name = []
product_name = []
prices = []
exchange = []
expect_delivery = []
availability = []
product_url=[]

for i in urls:
    
    driver.get(i)
    
    
    #fetching brand name 
    try:
        brand = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[20]/div[1]/div/div[1]/div[1]/div[3]/div[2]/div/span/a')
        brand_name.append(brand.text)
    except NoSuchElementException:
        brand_name.append('-')
    
    
    # fetching Name of the Product
    try:
        product = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[4]/div[1]/div/h1/span')
        product_name.append(product.text)
    except NoSuchElementException:
        product_name.append('-')
            
            
           

    #fetching price of the product    
    try:
        price =driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[1]/div/div')
        prices.append(price.text)
    except NoSuchElementException:
        prices.append('-')
        
        
    #fetching return/exchange
    try:
        exch = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[20]/div[1]/div/div[1]/div[1]/div[4]/div[2]/span/a/span')
        exchange.append(exch.text)
    except NoSuchElementException:
        exchange.append('-')
        

    #fetching expected delivery
    try:
        delivery = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[3]/div[10]/div[1]/div/div/div')
        expect_delivery.append(delivery.text)
    except NoSuchElementException:
        expect_delivery.append('-')
        

    #fetching availability information
    try:
        available = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[8]/div[3]/div[1]/div[5]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[6]/div/div[1]/span')
        availability.append(available.text)
    except NoSuchElementException:
        availability.append('-')
        
    #url details
    try:
        url = driver.find_element(By.XPATH,'//a[@class="ssf-share-trigger"]')
        product_url.append(url.text)
    except NoSuchElementException:
        product_url.append('-')
```

```python
print(len(brand_name),
len(product_name),
len(prices),
len(exchange),
len(expect_delivery),
len(availability),
len(product_url))
```

```python
#7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped: 
#“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”. 
```

```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
import requests
from selenium.webdriver.common.by import By
from selenium import webdriver

# connecting to the webdriver
driver=webdriver.Chrome()

# opening the url track.in
driver.get("https://www.forbes.com/billionaires/")
```

```python
# scraping required data from the web page  #//div[@class="Table_rank___YBhk Table_dataCell__2QCve"]
# creating empty lists
Rank = []
Name = []
Net_worth = []
Age = []
Citizenship = []
Source = []
Industry = []

while(True):
    
    # scraping the data of rank of the billionaires
    rank_tag = driver.find_elements(By.XPATH,'//div[@class="Table_rank___YBhk Table_dataCell__2QCve"]')
    for rank in rank_tag:
        Rank.append(rank.text)
    
    
    
 
    # scraping the data  of names of the billionaires
    name_tag = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][2]')
    for name in name_tag:
        Name.append(name.text)
    
    
    
    # scraping the data of age of the billionaires
    age_tag = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][4]')
    for age in age_tag:
        Age.append(age.text)
    
    
    
    # scraping the data of citizenship of the billionaires
    cit_tag = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][5]')
    for cit in cit_tag:
        Citizenship.append(cit.text)
    
    
    
    # scraping the data of source of income of the billionaires
    sour_tag = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][6]')
    for sour in sour_tag:
        Source.append(sour.text)
    
    
    
    # scraping data of industry of the billionaires
    ind_tag = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][7]')
    for ind in ind_tag:
        Industry.append(ind.text)
    
    
    
    # scraping data of net_worth of billionaires
    net_tag = driver.find_elements(By.XPATH,'//div[@class="TableRow_cell__db-hv Table_cell__houv9"][3]')
    for net in net_tag:
        Net_worth.append(net.text)
    
    
    
    # clicking on next button
    try:
        next_button = driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div[2]/div[2]/div[2]/div[1]/button[2]/div")
        next_button.click()
    except:
        break
```

```python
print(len(Rank),
len(Name),
len(Net_worth),
len(Age),
len(Citizenship),
len(Source),
len(Industry))
```

```python
# Creating the DataFrame for the scraped data

df = pd.DataFrame({'Rank':Rank,'Name':Name,'Net_worth':Net_worth,'Age':Age,'Citizenship':Citizenship,'Source':Source,'Industry':Industry})
df
```

```python
#saving the data in csv
df.to_csv("df.csv")
```

```python
# finding element for search bar
search_bar = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')
search_bar.send_keys("London")
```

```python
urls = []          # empty list
for i in range(0,1):      
    page_url = driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
    for i in page_url:
        urls.append(i.get_attribute("href"))
        
```

```python
# creating empty list & find required data
hostel_name = []
distance = []
pvt_prices = []
dorms_price = []
rating = []
reviews = []
over_all = []
facilities = []
description = []
url = []
```

```python
# scraping the required informations
    
    
    # scraping  hostel name
try:
    name = driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
    for i in name:
        hostel_name.append(i.text)
except NoSuchElementException:
        hostel_name.append('-')
        
        
    # scraping distance from city centre
try:
    dist = driver.find_elements(By.XPATH,'//div[@class="property-distance"]/span[2]')
    for i in dist:
            distance.append(i.text)
except NoSuchElementException:
        distance.append('-')
        
```
